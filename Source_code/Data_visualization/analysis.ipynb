```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
control = [[5,5,5,5,5,5],
           [4,4,5,3,5,4],
           [3,3,5,3,5,5],
           [4,4,4,4,4,3],
           [3,1,2,3,3,3],
           [4,3,4,3,4,3],
           [4,4,5,4,4,4],
           [5,5,5,4,5,3]
           ]

humor = [[5,5,5,5,5,5],
         [4,4,5,2,5,4],
         [3,2,3,2,3,4],
         [4,4,4,4,4,3],
         [4,2,3,2,4,3],
         [3,3,4,3,3,3],
         [4,4,4,2,4,4],
         [5,4,5,2,5,3]
          ]

# Create the dataframes
df_humor = pd.DataFrame(humor, columns=['enjoyable', 'humanness', 'appropriate', 'humorous', 'understanding', 'trust'])
df_control = pd.DataFrame(control, columns=['enjoyable', 'humanness', 'appropriate', 'humorous', 'understanding', 'trust'])

# Compute the column-wise average for each condition
avg_humor = df_humor.mean()
avg_control = df_control.mean()

# Create a new Pandas DataFrame that combines the average values for each condition
df_avg = pd.DataFrame({'Humor': avg_humor, 'Control': avg_control})

# Plot the bar chart with two bars side by side
ax = df_avg.plot.bar(rot=0)
ax.set_xlabel('Question')
ax.set_ylabel('Average Score')
plt.show()
```
